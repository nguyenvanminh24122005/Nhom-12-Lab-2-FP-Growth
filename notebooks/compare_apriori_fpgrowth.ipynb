```python
import pandas as pd

basket_bool = pd.read_parquet("../data/basket_bool.parquet")

print(basket_bool.shape)
```

```python
from mlxtend.frequent_patterns import apriori
import time

start = time.time()

freq_ap = apriori(
    basket_bool,
    min_support=0.01,
    use_colnames=True
)

ap_time = time.time() - start

print("Apriori time (seconds):", ap_time)
print("Apriori frequent itemsets:", freq_ap.shape)
```

```python
from mlxtend.frequent_patterns import fpgrowth

start = time.time()

freq_fp = fpgrowth(
    basket_bool,
    min_support=0.01,
    use_colnames=True
)

fp_time = time.time() - start

print("FP-Growth time (seconds):", fp_time)
print("FP-Growth frequent itemsets:", freq_fp.shape)
```

## So sánh Apriori và FP-Growth

- FP-Growth có thời gian xử lý nhanh hơn Apriori trên cùng tập dữ liệu.
- Apriori tiêu tốn nhiều bộ nhớ do phải sinh tập ứng viên.
- FP-Growth phù hợp hơn với tập dữ liệu lớn như Online Retail.

Kết luận: FP-Growth hiệu quả và khả năng mở rộng tốt hơn Apriori.
