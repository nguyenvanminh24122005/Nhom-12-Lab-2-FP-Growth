# Basket Preparation

## Mục tiêu
- Chuyển dữ liệu giao dịch đã làm sạch thành dạng giỏ hàng (basket).
- Chuẩn bị dữ liệu dạng one-hot encoding để phục vụ thuật toán Apriori.


```python
import pandas as pd

DATA_PATH = "../data/cleaned_online_retail.parquet"

df_clean = pd.read_parquet(DATA_PATH)

print("Đã load dữ liệu sạch:", df_clean.shape)
```

```python
basket = (
    df_clean
    .groupby(["InvoiceNo", "Description"])["Quantity"]
    .sum()
    .unstack()
    .fillna(0)
)

print("Kích thước basket:", basket.shape)
basket.head()
```

```python
basket_bool = basket.applymap(lambda x: 1 if x > 0 else 0)

print("Basket boolean:", basket_bool.shape)
basket_bool.head()
```

```python
OUTPUT_PATH = "../data/basket_bool.parquet"

basket_bool.to_parquet(OUTPUT_PATH)

print("Đã lưu basket tại:", OUTPUT_PATH)
```
