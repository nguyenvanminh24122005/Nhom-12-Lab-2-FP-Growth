```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
DATA_PATH = "../data/basket_bool.parquet"

basket_bool = pd.read_parquet(DATA_PATH)
print("Shape basket:", basket_bool.shape)
```

```python
frequent_itemsets_fp = fpgrowth(
    basket_bool,
    min_support=0.02,
    use_colnames=True
)

print("Số frequent itemsets:", frequent_itemsets_fp.shape)
frequent_itemsets_fp.head()
```

```python
rules_fp = association_rules(
    frequent_itemsets_fp,
    metric="lift",
    min_threshold=1.0
)

print("Số luật FP-Growth:", rules_fp.shape)
rules_fp.head()
```

```python
rules_fp.to_csv("../data/fp_growth_rules.csv", index=False)
print("Đã lưu fp_growth_rules.csv")
```
